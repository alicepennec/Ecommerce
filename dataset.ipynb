```python
""" # 1. Création de la dimension Date avec DateID unique
dim_date = df[['Date']].drop_duplicates().reset_index(drop=True)
dim_date['DateID'] = dim_date.index + 1  # Crée un ID unique
df = df.merge(dim_date, on='Date', how='left')  # Ajoute DateID au DataFrame principal """

# 2. Création de la dimension Produit avec ProductID unique
dim_product = df[['ProductName', 'ProductCategory', 'ProductSubCategory', 'ProductPrice']].drop_duplicates().reset_index(drop=True)
dim_product['ProductID'] = dim_product.index + 1
df = df.merge(dim_product, on=['ProductName', 'ProductCategory', 'ProductSubCategory', 'ProductPrice'], how='left')

# 3. Création de la dimension Client avec CustomerID unique
dim_customer = df[['CustomerName', 'CustomerEmail', 'CustomerAddress', 'CustomerPhone', 'CustomerSegment']].drop_duplicates().reset_index(drop=True)
dim_customer['CustomerID'] = dim_customer.index + 1
df = df.merge(dim_customer, on=['CustomerName', 'CustomerEmail', 'CustomerAddress', 'CustomerPhone', 'CustomerSegment'], how='left')

# 4. Création de la dimension Fournisseur avec SupplierID unique
dim_supplier = df[['SupplierName', 'SupplierLocation', 'SupplierContact']].drop_duplicates().reset_index(drop=True)
dim_supplier['SupplierID'] = dim_supplier.index + 1
df = df.merge(dim_supplier, on=['SupplierName', 'SupplierLocation', 'SupplierContact'], how='left')

# 5. Création de la dimension Expédition avec ShipperID unique
dim_shipper = df[['ShipperName', 'ShippingMethod']].drop_duplicates().reset_index(drop=True)
dim_shipper['ShipperID'] = dim_shipper.index + 1
df = df.merge(dim_shipper, on=['ShipperName', 'ShippingMethod'], how='left')

# 6. Création de la table de faits SalesFact en utilisant les clés ID générées
sales_fact = df[['Date', 'ProductID', 'CustomerID', 'SupplierID', 'ShipperID', 'QuantitySold', 'TotalAmount', 'DiscountAmount', 'NetAmount']]

# 7. Création de la table de faits InventoryFact en utilisant les clés ID générées
inventory_fact = df[['Date', 'ProductID', 'StockReceived', 'StockSold', 'StockOnHand']]

df
```

## Importation des données dans MySQL

```python
import mysql.connector as MC

# création de la database
try:
    conn = MC.connect(host = 'localhost', port = '3306', user = 'root', password = 'root')
    cursor = conn.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS ecommerce")
except MC.Error as err:
    print(err)
```

```python
# création des tables de dimension
conn = MC.connect(host = 'localhost', port = '3306', user = 'root', password = 'root', database = "ecommerce")
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS dimproduct (
               ProductID INTEGER PRIMARY KEY,
               ProductName Varchar(50),
               ProductCategory Varchar(50),
               ProductSubCategory Varchar(50),
               ProductPrice Float
               );''')
conn.commit()
```

```python
# création des tables de dimension
conn = MC.connect(host = 'localhost', port = '3306', user = 'root', password = 'root', database = "ecommerce")
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS dimsupplier (
               SupplierID INTEGER PRIMARY KEY,
               SupplierName Varchar(50),
               SupplierLocation Varchar(50),
               SupplierContact Varchar(50)
               );''')
conn.commit()
```

```python
# création des tables de dimension
conn = MC.connect(host = 'localhost', port = '3306', user = 'root', password = 'root', database = "ecommerce")
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS dimshipper (
               ShipperID INTEGER PRIMARY KEY,
               ShipperName Varchar(50),
               ShippingMethod Varchar(50)
               );''')
conn.commit()
```

```python
# création des tables de faits
conn = MC.connect(host = 'localhost', port = '3306', user = 'root', password = 'root', database = "ecommerce")
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS salesfact (
                    Date datetime,
                    ProductID integer,	
                    CustomerID integer,
                    SupplierID integer,
                    ShipperID integer,	
                    QuantitySold integer,	
                    TotalAmount	float,
                    DiscountAmount float,
                    NetAmount float,
                    PRIMARY KEY (Date),
                    CONSTRAINT fk_product FOREIGN KEY (ProductID) REFERENCES dimproduct (ProductID),
                    CONSTRAINT fk_supplier FOREIGN KEY (SupplierID) REFERENCES dimsupplier (SupplierID),
                    CONSTRAINT fk_customer FOREIGN KEY (CustomerID) REFERENCES dimcustomer (CustomerID),
                    CONSTRAINT fk_shipper FOREIGN KEY (ShipperID) REFERENCES dimshipper (ShipperID)
               );''')
conn.commit()
```

```python
# création des tables de faits
conn = MC.connect(host = 'localhost', port = '3306', user = 'root', password = 'root', database = "ecommerce")
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS inventoryfact (
                    Date datetime,
                    ProductID integer,
                    StockReceived integer,
                    StockSold integer,
                    StockOnHand integer,
                    PRIMARY KEY (Date),
                    CONSTRAINT fk_product2 FOREIGN KEY (ProductID) REFERENCES dimproduct (ProductID)
               );''')
conn.commit()
```

### Insertion des données

```python
for i, row in dim_product.iterrows():
    cursor.execute("""
        INSERT INTO dimproduct (ProductID, ProductName, ProductCategory, ProductSubCategory, ProductPrice)
        VALUES (%s, %s, %s, %s, %s)
    """, (row['ProductID'], row['ProductName'], row['ProductCategory'], row['ProductSubCategory'], row['ProductPrice']))

conn.commit()
```

```python
# Insertion des données dans la table dimcustomer
for i, row in dim_customer.iterrows():
    cursor.execute("""
        INSERT INTO dimcustomer (CustomerID, CustomerName, CustomerEmail, CustomerAddress, CustomerPhone, CustomerSegment)
        VALUES (%s, %s, %s, %s, %s, %s)
    """, (row['CustomerID'], row['CustomerName'], row['CustomerEmail'], row['CustomerAddress'], row['CustomerPhone'], row['CustomerSegment']))
conn.commit()
```

```python
# Insertion des données dans la table dimsupplier
for i, row in dim_supplier.iterrows():
    cursor.execute("""
        INSERT INTO dimsupplier (SupplierID, SupplierName, SupplierLocation, SupplierContact)
        VALUES (%s, %s, %s, %s)
    """, (row['SupplierID'], row['SupplierName'], row['SupplierLocation'], row['SupplierContact']))

conn.commit()
```

```python
# Insertion des données dans la table dimshipper
for i, row in dim_shipper.iterrows():
    cursor.execute("""
        INSERT INTO dimshipper (ShipperID, ShipperName, ShippingMethod)
        VALUES (%s, %s, %s)
    """, (row['ShipperID'], row['ShipperName'], row['ShippingMethod']))

conn.commit()
```

```python
# Utilisation de executemany pour insérer plusieurs lignes à la fois

data_to_insert = list(inventory_fact_clean.itertuples(index=False, name=None))
cursor.executemany("""
    INSERT INTO inventoryfact (Date, ProductID, StockReceived, StockSold, StockOnHand)
    VALUES (%s, %s, %s, %s, %s)
    ON DUPLICATE KEY UPDATE 
    StockReceived = VALUES(StockReceived),
    StockSold = VALUES(StockSold),
    StockOnHand = VALUES(StockOnHand)
""", data_to_insert)

# Commit des changements
conn.commit()
```
